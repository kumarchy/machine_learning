 # checking whether the price of canceled or not canceled reservation is high

```python
canceled_df_adr=canceled_data.groupby('reservation_status_date')[['adr']].mean()
canceled_df_adr.reset_index(inplace=True)
canceled_df_adr.sort_values('reservation_status_date',inplace=True)

not_canceled_data=df[df['is_canceled']==0]
not_canceled_df=not_canceled_data.groupby('reservation_status_date')[['adr']].mean()
not_canceled_df.reset_index(inplace=True)
not_canceled_df.sort_values('reservation_status_date',inplace=True)

plt.figure(figsize=(20,10))
plt.title('Average Daily Rate')
plt.plot(not_canceled_df['reservation_status_date'],not_canceled_df['adr'],label='not canceled')
plt.plot(canceled_df_adr['reservation_status_date'],canceled_df_adr['adr'],label='canceled')
plt.legend()
```

# data between 2016 to 2017

```python
canceled_df_adr=canceled_df_adr[(canceled_df_adr['reservation_status_date']>'2016') & (canceled_df_adr['reservation_status_date']<'2017-09')]
not_canceled_df=not_canceled_df[(not_canceled_df['reservation_status_date']>'2016') & (not_canceled_df['reservation_status_date']<'2017-09')]
```

```python
plt.figure(figsize=(20,8))
plt.title('Averag Daily Rate',size=30)
plt.plot(not_canceled_df['reservation_status_date'],not_canceled_df['adr'],label='not canceled')
plt.plot(canceled_df_adr['reservation_status_date'],canceled_df_adr['adr'],label='canceled')

plt.legend()
```
