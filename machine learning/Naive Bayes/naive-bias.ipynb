```python
# loading the data
df=pd.read_csv('/kaggle/input/play-tennis/play_tennis (1).csv')
df.head()
```

```python
# droping the columns
df.drop(columns=['day'],inplace=True)
```

```python
df.head()
```

```python
# splitting the data for x and y
x=df.drop(columns=['play'])
y=df.play.ravel()
y
```

```python
# one hot encoding
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(drop='first',sparse=False,dtype=np.int32)

x= ohe.fit_transform(x[['outlook','temp','humidity','wind']])
x
```

```python
print(y)

# label encoding
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
y=le.fit_transform(y)
y
```

```python
# importing the model
from sklearn.naive_bayes import CategoricalNB
from sklearn.metrics import accuracy_score
```

```python
nb_classifier=CategoricalNB()

# model_fitting
nb_classifier.fit(x,y)
```

```python
# model_prediction
y_pred = nb_classifier.predict(x)
y_pred
```

```python
# accuracy
print(accuracy_score(y,y_pred))
```

```python
# prediction on new_data
x1 = np.array([0, 1, 1, 0, 0, 0]).reshape(1, -1)
nb_classifier.predict(x1)
```
