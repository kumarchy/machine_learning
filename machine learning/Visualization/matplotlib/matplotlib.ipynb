# Piechart

```python
areas=['marketing','sales','development','HR','customer support']
budget=[2.5,4,10,1,6]
```

```python
x=[23,45,12,49,97,32,11,44]
y=[45,34,67,87,55,22,89,90]
plt.scatter(x,y)
plt.xlabel('age')
plt.ylabel('salary')
plt.title('age vs salary graph')
```

```python
product_id=[1,2,3,4,5,6,7,8]
online_price=[233,456,770,120,222,444,200,300]
```

```python
plt.plot(product_id,online_price,color='green',marker='D',markersize=10)
plt.xlabel('product')
plt.ylabel('prices')
plt.title('product price comparision')
plt.show()
```

```python
product_id=[1,2,3,4,5,6,7,8]
online_price=[233,456,770,120,222,444,200,300]
offline_price=[400,300,100,333,444,565,565,899]
chor_bazaar=[100,200,300,400,150,250,450,500]
```

```python
plt.plot(product_id,online_price,color='green',marker='D',markersize=10,label='online prices')
plt.plot(product_id,offline_price,color='yellow',markersize=10,label='offline prices')
plt.plot(product_id,chor_bazaar,color='red',marker='+',markersize=10,label='chori prices')
plt.xlabel('product')
plt.ylabel('prices')
plt.title('product price comparision')
plt.legend()
plt.show()
```

```python
year=[2012,2013,2014,2015,2016,2018]
price=[65000,69000,75000,55000,64000,84000]
plt.bar(year,price,color='red')
plt.xlabel('year')
plt.ylabel('price(in INR)')
plt.title('year by year price comparision')
plt.show()
```
