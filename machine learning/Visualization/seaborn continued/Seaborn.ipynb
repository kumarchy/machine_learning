# Heatmap

```python
flight=pd.DataFrame({'year':[1949,1949,1949,1949,1949,1948,1948,1948,1946,1946,1946,1945],
                'month':['january','february','march','april','may','june','july','august','september','october','november','december'],
                 'passenger':[112,118,200,125,136,145,120,218,254,110,116,117]})
```

```python
flight
```

```python
#sample only
x=flight.pivot_table(index='year',columns='month',values='passenger',aggfunc='sum')
```

```python
sns.heatmap(x,cbar=False)
```

```python
# we can fill the all box
#sns.heatmap(x,cbar=False,linewidth=0.5,annot=True,fmt='d')
```

```python
sns.heatmap(x,cbar=False,linewidth=0.5,annot=True,cmap='summer')
```

# distplots

# distplot combines 3 things in one graph
## histogram
## kernel density estimation(pdf)
## Rugplot

```python
titanic=sns.load_dataset('titanic')
titanic.head()
```

```python
titanic['age'].fillna(titanic['age'].mean(),inplace=True)
```

```python
sns.distplot(titanic['age'])
```

```python
# only Histogram
sns.distplot(titanic['age'],bins=10,kde=False)
```

```python
#only kde
sns.distplot(titanic['age'],hist=False)
```

```python
# it is actually histogram
sns.distplot(titanic['age'],hist=False,kde=False,rug=True)
```

```python
sns.distplot(titanic[titanic['survived']==1]['age'])
sns.distplot(titanic[titanic['survived']==0]['age'])
```
